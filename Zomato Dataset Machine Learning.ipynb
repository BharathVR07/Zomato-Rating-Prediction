```python
top_10 = [x for x in final_df['City'].value_counts().sort_values(ascending=False).head(10).index]
top_10
```

```python
for label in top_10:
    final_df[label] = np.where(final_df['City']==label,1,0)
```

```python
top_10 = [x for x in final_df['Cuisines'].value_counts().sort_values(ascending=False).head(10).index]
print(top_10)
for label in top_10:
    final_df[label] = np.where(final_df['Cuisines']==label,1,0)
```

```python
final_df.columns
```

```python
currency = pd.get_dummies(final_df['Currency'])
currency
```

```python
final_df = pd.concat([final_df, currency], axis=1)
```

```python
final_df.drop('Currency', axis=1, inplace=True)
```

```python
final_df['Has Table booking'].unique()
```

```python
final_df['Has Table booking'] = pd.get_dummies(final_df['Has Table booking'], drop_first=True)
```

```python
final_df['Has Online delivery'].unique()
```

```python
final_df['Has Online delivery'] = pd.get_dummies(final_df['Has Online delivery'], drop_first=True)
```

```python
final_df['Is delivering now'].unique()
```

```python
final_df['Is delivering now'] = pd.get_dummies(final_df['Is delivering now'], drop_first=True)
```

```python
final_df.info()
```

```python
final_df['Rating color'] = final_df['Rating color'].map({'Dark Green':1,'Green':2,'Yellow':3,'Orange':4,'Red':5,'White':6})
```

```python
final_df['Rating text'] = final_df['Rating text'].map({'Excellent':1,'Very Good':2,'Good':3,'Average':4,'Poor':5,'Not rated':6})
```

```python
top_10 = [x for x in final_df['Country'].value_counts().sort_values(ascending=False).head(10).index]
for label in top_10:
    final_df[label] = np.where(final_df['Country']==label,1,0)
```

```python
top_10 = [x for x in final_df['Locality'].value_counts().sort_values(ascending=False).head(10).index]
for label in top_10:
    final_df[label] = np.where(final_df['Locality']==label,1,0)
```

```python
final_df.info()
```

```python
final_df.drop(['City','Locality','Cuisines','Country'], axis=1, inplace=True)
```

```python
final_df.columns
```

```python
final_df.info()
```

```python
X = final_df.drop('Aggregate rating', axis=1)
y = final_df['Aggregate rating']
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.8, random_state=42)
```

## Ordinary Least Squared Method.

```python
import statsmodels.api as sm
X = sm.add_constant(X)
result = sm.OLS(y, X).fit()
print(result.summary())
```

```python
from sklearn.ensemble import ExtraTreesRegressor
selection = ExtraTreesRegressor()
selection.fit(X_train, y_train)
```

```python
selected_feat = pd.Series(selection.feature_importances_, index=X_train.columns)
selected_feat.nlargest(10).plot(kind='barh')
plt.show()
```

```python
features_selected = selected_feat.nlargest(10).index
features_selected
```

```python
X_train = X_train[features_selected]
X_test = X_test[features_selected]
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

## Linear Regression.

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
```

```python
lin_reg.fit(X_train,y_train)
```

```python
lin_reg.score(X_train,y_train)
```

```python
y_predict = lin_reg.predict(X_test)
```

```python
lin_reg.score(X_test,y_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_predict)
```

# R2 Score for Linear Regression is 0.90

## Ridge Regression.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.8, random_state=42)
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
ridge = Ridge()
params = {'alpha':[1e-15,1e-13,1e-12, 1e-6,1e-3,1e-2,1,3,17,21,54,108]}
ridge_regressor = GridSearchCV(ridge, params, scoring='neg_mean_squared_error', cv=6)
ridge_regressor.fit(X_train, y_train)
```

```python
y_pred_ridge = ridge_regressor.predict(X_test)
```

```python
r2_score(y_test, y_pred_ridge)
```

# R2 Score for Rdige Regression is 0.90

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
lasso = Lasso()
params = {'alpha':[1e-15,1e-13,1e-12, 1e-6,1e-3,1e-2,1,3,17,21,54,36,47,31,7]}
lasso_regressor = GridSearchCV(ridge, params, scoring='neg_mean_squared_error', cv=6)
lasso_regressor.fit(X_train, y_train)
```

```python
y_pred_lasso = lasso_regressor.predict(X_test)
```

```python
r2_score(y_test, y_pred_lasso)
```

# R2 Score for Lasso Regression is 0.90
